1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
#ваш код здесь
def extract_education_level(edu_info: str) -> str:
    '''
    Извлекает уровень образования:
    "Высшее образование 1987 Военный инженерный..."  --> "высшее"
    '''
    #  не будем проверять на NaN, df.info() говорит, что колонка заполнена полностью
    
    terms = edu_info.lower().split()
    if terms[0] == 'высшее':
        return terms[0]
    elif terms[0] == 'неоконченное':
        return 'неоконченное высшее'
    
    return 'среднее специальное' if terms[1] == 'специальное' else 'среднее'


data['Образование'] = data['Образование и ВУЗ'].apply(extract_education_level).astype('category')
data.drop(columns='Образование и ВУЗ', inplace=True)
display(data['Образование'].value_counts())
```

2. Теперь нас интересует столбец **"Пол, возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:
* Мужчина , 39 лет , родился 27 ноября 1979 
* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
#ваш код здесь
def extract_age(value: str) -> int:
    '''
    Выделяет возраст из строки вида "Мужчина , 39 лет , родился 27 ноября 1979".
    Возвращает целое число - 39 в данном случае
    '''

    terms = value.split(',')
    terms = terms[1].strip().split()
    return int(terms[0])


data['Пол'] = data['Пол, возраст'].apply(lambda x: 'М' if x.startswith('Муж') else 'Ж')\
                .astype('category')
data['Возраст'] = data['Пол, возраст'].apply(extract_age).astype('int8')
data.drop(columns='Пол, возраст', inplace=True)

# Сколько процентов женских резюме представлено в наших данных? Ответ округлите до сотых.
val = data['Пол'].value_counts(normalize=True)['Ж']
print('Процентов женских резюме: {}'.format(round(100 * val, 2)))

# Чему равен средний возраст соискателей? Ответ округлите до десятых.
print('Средний возраст соискателей: {}'.format(round(data['Возраст'].mean(), 1))) 
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь
def extract_experience_months(value: str) -> int:
    '''
    Из строки вида "Опыт работы 15 лет 10 месяцев Руководитель..." вычисляет опыт в месяцах:
    - Опыт работы 15 лет 10 месяцев --> 190
    - Опыт работы 10 месяцев --> 10
    - Опыт работы 5 лет --> 60
    - Не указано --> nan
    '''
    
    if not value:
        return np.nan
    if value is np.nan:
        return np.nan
    
    value = value.lower()
    if value.startswith('не указан'):
        return np.nan
    
    terms = value.split()
    # если после "Опыт работы" не число, вернем NaN
    if not terms[2].isdecimal():
        return np.nan
    
    res = int(terms[2])
    # если после числа идет месяц(ы,ев), вернем значение
    if terms[3].startswith('месяц'):
        return res
    
    # сюда попадаем только если первыми идут года
    res *= 12
    if terms[4].isdecimal():
        res += int(terms[4])
    return res


data['Опыт работы (месяц)'] = data['Опыт работы'].apply(extract_experience_months)
data.drop(columns=['Опыт работы'], inplace=True)

# Чему равен медианный опыт работы (в месяцах) в нашей таблице?
print('Медианный опыт работы (в месяцах): {}'.format(data['Опыт работы (месяц)'].median()))
```

    * … , готов к командировкам , …

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
def extract_city(city: str)->str:
    '''
    Выделяет признак «Город» из строки вида "Тверь , не готова к переезду , не готова к ком..."
    Может содержать только четыре категории: «Москва», «Санкт-Петербург», «город-миллионник» (их список ниже) и «другие».
    '''
    main_cities = ['Москва', 'Санкт-Петербург']
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск', 'Омск', 
                      'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    if city in main_cities:
        return city
    if city in million_cities:
        return 'город-миллионник'
    return 'другие'


def extract_flags(value: str):
    '''
    Выделяет признаки "Готов к переезду" и "Готовность к командировкам" из строки вида "Тверь , не готова к переезду , не готова к ком..."
    Вернет пару вида [True, False]
    '''
    
    terms = value.split(' , ')
    city = extract_city(terms[0])

    res = []
    for tmp in terms[1:]:
        tmp = tmp.strip()
        if tmp.startswith('готов') or tmp.startswith('хочу'):
            res.append(True)
        if tmp.startswith('не готов'):
            res.append(False)
    
    if len(res) == 1:
        res.append(False) # потерянный "Не готов(а) к командировкам"
    
    return pd.Series({'Город': city, 'Готовность к переезду':res[0], 'Готовность к командировкам': res[1]})


# создадим Dataframe для Город, Готовность к переезду, Готовность к командировкам
df = data['Город, переезд, командировки'].apply(extract_flags)

# выполним merge с основым Dataframe и поправим типы колонок
data_merged = data.merge(df, left_index=True, right_index=True)
data_merged['Город'] = data_merged['Город'].astype('category')
data_merged['Готовность к переезду'] = data_merged['Готовность к переезду'].astype('bool')
data_merged['Готовность к командировкам'] = data_merged['Готовность к командировкам'].astype('bool')
data_merged.drop(columns='Город, переезд, командировки', inplace=True)
```

```python
# Сколько процентов соискателей живут в Санкт-Петербурге? Ответ округлите до целого.
print("в Санкт-Петербурге живут {}% соискателей".format(round(100 * data_merged['Город'].value_counts(normalize=True)['Санкт-Петербург'])))

# Сколько процентов соискателей готовы одновременно и к переездам, и к командировкам? Ответ округлите до целого.
df = data_merged[data_merged['Готовность к переезду'] & data_merged['Готовность к командировкам']]
print("{}% соискателей готовы одновременно и к переездам, и к командировкам".format(round(100 * df.shape[0] / data_merged.shape[0])))
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь
def extract_unique_values(column_data: pd.Series) -> set:
    '''
    Метод создает set уникальных значений (строк) для переданной колонки.
    Пример: допустим у нас есть 2 строки в переданном Series
        - "полная занятость, частичная занятость"
        - "частичная занятость, проектная работа, волонтерство"
        
    на выходе мы ожидаем set значений: {'полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство'}
    '''
    
    res = set()
    for str in column_data:
        terms = str.split(',')
        for t in terms: res.add(t.strip())
    return res


# извлечем все типы занятости. это будут имена колонок с признаками-мигалками
work_types = extract_unique_values(data_merged['Занятость'])
# вычислим значения колонок
for work_type in work_types:
    data_merged[work_type] = data_merged['Занятость'].apply(lambda x: str(x).find(work_type) >= 0)
    
# повторим для "График"        
schedule_types = extract_unique_values(data_merged['График'])
for schedule_type in schedule_types:
    data_merged[schedule_type] = data_merged['График'].apply(lambda x: str(x).find(schedule_type) >= 0)
    
data_merged.drop(columns=['Занятость', 'График'], inplace=True)
```

```python
# Сколько людей ищут проектную работу и волонтерство (в обоих столбцах стоит True)?
val = data_merged[data_merged['проектная работа'] & data_merged['волонтерство']].shape[0]
print('{} людей ищут проектную работу и волонтерство одновременно'.format(val))

# Сколько людей хотят работать вахтовым методом и с гибким графиком (в обоих столбцах стоит True)?
val = data_merged[data_merged['вахтовый метод'] & data_merged['гибкий график']].shape[0]
print('{} людей хотят работать вахтовым методом и с гибким графиком'.format(val))
```

Например, для казахстанского тенге курс на 20.08.2019 составляет 17.197 руб. за 100 тенге, тогда итоговый курс равен - 17.197 / 100 = 0.17197 руб за 1 тенге.

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь

def extract_currency_ISO_code(value: str)->str:
    '''
    Извлекает ISO код валюты из строки:
        - "29000 руб." извлечем валюту "руб." и преобразуем в ISO код "RUB"
        - "50000 грн." извлечем валюту "грн." и преобразуем в ISO код "UAH"
    '''
    
    # замапим только коды, требующие преобразования
    iso_codes_dict = {
        'руб': 'RUB',
        'грн': 'UAH',
        'белруб': 'BYN',
        'сум': 'UZS'
    }
    terms = value.split()
    # на случай, если валюта не указана, будем считать что рубли
    if terms[0].strip().isdecimal() and len(terms) == 1:
        return 'RUB'
    tmp = terms[1].strip()
    tmp = terms[1].strip().replace('.','')
    if not iso_codes_dict.get(tmp):
        return tmp
    return  iso_codes_dict[tmp]


def extract_amount(value: str)->float:
    terms = value.split()
    return float(terms[0].strip())


data_merged['ЗП(Валюта)'] = data_merged['ЗП'].apply(extract_currency_ISO_code).astype('category')
data_merged['ЗП(Сумма)'] = data_merged['ЗП'].apply(extract_amount)
data_merged['Обновление резюме'] = pd.to_datetime(data_merged['Обновление резюме'], dayfirst=True).dt.date

# прочитаем курсы валют
rates = pd.read_csv('data/ExchangeRates.csv', sep=',')
# useless columns/data
rates.drop(columns=['per', 'time', 'vol'], inplace=True)
rates['currency'] = rates['currency'].astype('category') 
rates['date'] = pd.to_datetime(rates['date'], format='%d/%m/%y').dt.date

# объеденим таблицы
data_merged_with_rates = data_merged.merge(right=rates, how="left", left_on=['ЗП(Валюта)','Обновление резюме'], right_on=['currency', 'date'])
if data_merged_with_rates.shape[0] != data_merged.shape[0]:
    raise ValueError("incorrectly merged - number of rows changed!!!")

# проверим, что NaN для 'close' и 'proportion'  встречаются ТОЛЬКО для валюты "RUB" 
rows = data_merged_with_rates[(data_merged_with_rates['close'].isna() | data_merged_with_rates['proportion'].isna()) 
                              & (data_merged_with_rates['ЗП(Валюта)'] != 'RUB')].shape[0]
print('кол-во рядов с валютой != RUB и NaN в одной из колонок close и proportion', rows)
if rows > 0:
    raise ValueError("not all currencies can be converted to RUB")

# заполним пропуски единицами
data_merged_with_rates.fillna({'proportion': 1, 'close': 1}, inplace=True)

# пересчитаем ЗП в рубли
data_merged_with_rates['ЗП (руб)'] = data_merged_with_rates.apply(lambda row: round(row['ЗП(Сумма)'] * row['close'] / row['proportion'], 2), axis=1)

# удалим ненужные колонки
data_merged_with_rates.drop(columns=['ЗП', 'ЗП(Валюта)', 'ЗП(Сумма)', 'currency', 'date', 'close', 'proportion'], inplace=True)
display(data_merged_with_rates.info())
```

```python
def clear_z(data: pd.DataFrame, column_name: str, log: bool = False, left=3.0, right=3.0):
    x = data[column_name]
    if log:
        x = np.log(x)
    
    mean, std = x.mean(), x.std()
    lower_bound = mean - left * std
    upper_bound = mean + right * std
    
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    
    return cleaned, outliers


data_cleaned, outliers = clear_z(data_no_duplicates, 'Возраст', right=4.0)
display(outliers['Возраст'].describe())
```
